```python
def get_top_articles(n, df = df):
    '''
    INPUT:
    n - (int) the number of top articles to return
    df - (pandas dataframe) df as defined at the top of the notebook 
    
    OUTPUT:
    top_articles - (list) A list of the top 'n' article titles 
    
    '''
    # Your code here
    
    #counting number of times each article was interacted with
    Article_count = pd.DataFrame(df.groupby(['article_id', 'title']).size().reset_index(name = 'Count'))
    
    #sorting articles count in descending order
    sorte = Article_count.sort_values('Count', ascending = False)
    
    #selecting required top articles
    top_articles =  list(sorte['title'][:n])
    
    return top_articles # Return the top article titles from df (not df_content)

def get_top_article_ids(n, df=df):
    '''
    INPUT:
    n - (int) the number of top articles to return
    df - (pandas dataframe) df as defined at the top of the notebook 
    
    OUTPUT:
    top_articles - (list) A list of the top 'n' article titles 
    
    '''
    # Your code here
    
    #same code above but for article ids
    
    Article_count = pd.DataFrame(df.groupby(['article_id', 'title']).size().reset_index(name = 'Count'))
    
    sorte = Article_count.sort_values('Count', ascending = False)

    top_articles =  list(sorte['article_id'][:n])
 
    return top_articles # Return the top article ids
```

### <a class="anchor" id="User-User">Part III: User-User Based Collaborative Filtering</a>


`1.` Use the function below to reformat the **df** dataframe to be shaped with users as the rows and articles as the columns.  

* Each **user** should only appear in each **row** once.


* Each **article** should only show up in one **column**.  


* **If a user has interacted with an article, then place a 1 where the user-row meets for that article-column**.  It does not matter how many times a user has interacted with the article, all entries where a user has interacted with an article should be a 1.  


* **If a user has not interacted with an item, then place a zero where the user-row meets for that article-column**. 

Use the tests to make sure the basic structure of your matrix matches what is expected by the solution.

```python
# create the user-article matrix with 1's and 0's

def create_user_item_matrix(df):
    '''
    INPUT:
    df - pandas dataframe with article_id, title, user_id columns
    
    OUTPUT:
    user_item - user item matrix 
    
    Description:
    Return a matrix with user ids as rows and article ids on the columns with 1 values where a user interacted with 
    an article and a 0 otherwise
    '''
    # Fill in the function here
    
    #creating a user-item matrix of user, article id with 1 for interaction, 0 for no interaction
    
    user_articles = df[['user_id', 'article_id']]
    
    user_articles = user_articles.loc[~user_articles.duplicated(keep = 'first')]

    user_articles['interactions'] = np.ones(len(user_articles))

    user_item = user_articles.groupby(['user_id', 'article_id'])['interactions'].max().unstack()
    
    user_item = user_item.replace(np.nan, 0)
    
    
    
    return user_item # return the user_item matrix 


#saving created user item matrix
import pickle
user_item = create_user_item_matrix(df)
f = open("user_item_matrix.pkl", "wb")
pickle.dump(user_item, f)
f.close()
```

`2.` Complete the function below which should take a user_id and provide an ordered list of the most similar users to that user (from most similar to least similar).  The returned result should not contain the provided user_id, as we know that each user is similar to him/herself. Because the results for each user here are binary, it (perhaps) makes sense to compute similarity as the dot product of two users. 

Use the tests to test your function.

```python
def find_similar_users(user_id, user_item=user_item):
    '''
    INPUT:
    user_id - (int) a user_id
    user_item - (pandas dataframe) matrix of users by articles: 
                1's when a user has interacted with an article, 0 otherwise
    
    OUTPUT:
    similar_users - (list) an ordered list where the closest users (largest dot product users)
                    are listed first
    
    Description:
    Computes the similarity of every pair of users based on the dot product
    Returns an ordered
    
    '''
    # compute similarity of each user to the provided user
    
    #fnding most similar using dot product and returning list of ids
    
    u1 = np.array(user_item.iloc[user_id - 1, :])
    users = np.array(user_item.iloc[:, :])
    dotprod = np.dot(u1, users.T)

    # sort by similarity
    dotsort = (dotprod.argsort()[::-1] + 1)
    
    # create list of just the ids
    
    id_list = list(dotsort)
    
    # remove the own user's id
    id_list.remove(user_id)
    
    most_similar_users = id_list
    
       
    return most_similar_users # return a list of the users in order from most to least similar
        
```

`3.` Now that you have a function that provides the most similar users to each user, you will want to use these users to find articles you can recommend.  Complete the functions below to return the articles you would recommend to each user. 

```python
def get_article_names(article_ids, df=df):
    '''
    INPUT:
    article_ids - (list) a list of article ids
    df - (pandas dataframe) df as defined at the top of the notebook
    
    OUTPUT:
    article_names - (list) a list of article names associated with the list of article ids 
                    (this is identified by the title column)
    '''
    # Your code here
    #returning article names for recommendations
    article_names = list(df['title'].loc[df['article_id'].isin (article_ids)].unique())
    
    return article_names # Return the article names associated with list of article ids


def get_user_articles(user_id, user_item=user_item):
    '''
    INPUT:
    user_id - (int) a user id
    user_item - (pandas dataframe) matrix of users by articles: 
                1's when a user has interacted with an article, 0 otherwise
    
    OUTPUT:
    article_ids - (list) a list of the article ids seen by the user
    article_names - (list) a list of article names associated with the list of article ids 
    
    Description:
    Provides a list of the article_ids and article titles that have been seen by a user
    '''
    # Your code here
    
    #selecting articles already interacted with by specified user to ensure our recommendations are relevant
    
    article_ids = list((user_item.iloc[user_id - 1, :].loc[user_item.iloc[user_id -1, :] == 1]).index.values)
    
    article_names = get_article_names(article_ids, df=df)
    
    return article_ids, article_names # return the ids and names


def user_user_recs(user_id, m=10):
    '''
    INPUT:
    user_id - (int) a user id
    m - (int) the number of recommendations you want for the user
    
    OUTPUT:
    recs - (list) a list of recommendations for the user
    
    Description:
    Loops through the users based on closeness to the input user_id
    For each user - finds articles the user hasn't seen before and provides them as recs
    Does this until m recommendations are found
    
    Notes:
    Users who are the same closeness are chosen arbitrarily as the 'next' user
    
    For the user where the number of recommended articles starts below m 
    and ends exceeding m, the last items are chosen arbitrarily
    
    '''
    # Your code here

    recs = []
    
    #check artciles seen for specific user to avoid predicting same item already viewed
    articles_ids_seen, _ = get_user_articles(user_id, user_item)
    
    #Retrieving list of closest neighbors for specific user id
    closest_neighbors = find_similar_users(user_id)
    
    
    #loop through each all closest neighbour and use their interactions to make recommendations
    
    for neighbor in closest_neighbors:
        
        article_ids, _ = get_user_articles(neighbor, user_item=user_item)
        
        new_recs_ids = np.setdiff1d(np.array(article_ids), np.array(articles_ids_seen), assume_unique=True)
        
        
        #add recommendations from each user
        recs = np.concatenate([new_recs_ids, recs], axis=0)
        
        _, idx =np.unique(recs, return_index = True)
        
        recs = recs[np.sort(idx)]
        
        recs = list(recs)
        
        #break loop once a certain threshold of recommendations is reached
        if len(recs) > m:
            
            break
        
    
    return recs[:m] # return your recommendations for this user_id    
```

`4.` Now we are going to improve the consistency of the **user_user_recs** function from above.  

* Instead of arbitrarily choosing when we obtain users who are all the same closeness to a given user - choose the users that have the most total article interactions before choosing those with fewer article interactions.


* Instead of arbitrarily choosing articles from the user where the number of recommended articles starts below m and ends exceeding m, choose articles with the articles with the most total interactions before choosing those with fewer total interactions. This ranking should be  what would be obtained from the **top_articles** function you wrote earlier.

```python
def get_top_sorted_users(user_id, df=df, user_item=user_item):
    '''
    INPUT:
    user_id - (int)
    df - (pandas dataframe) df as defined at the top of the notebook 
    user_item - (pandas dataframe) matrix of users by articles: 
            1's when a user has interacted with an article, 0 otherwise
    
            
    OUTPUT:
    neighbors_df - (pandas dataframe) a dataframe with:
                    neighbor_id - is a neighbor user_id
                    similarity - measure of the similarity of each user to the provided user_id
                    num_interactions - the number of articles viewed by the user - if a u
                    
    Other Details - sort the neighbors_df by the similarity and then by number of interactions where 
                    highest of each is higher in the dataframe
     
    '''
    # Your code here
    
    #This recommendations would use both metrics of closest similarity along with number of interactions.
    
    #We have to create new code for finding neighbours to return the scores for each neighbour
    
    u1 = np.array(user_item.iloc[user_id - 1, :])
    
    users = np.array(user_item.iloc[:, :])
    
    user_similarity_scores = list(np.dot(u1, users.T))
    
    
    
    user_index = list(user_item.index)
    
    
    
    #This code creates a dataframe of neighbours, similarity and neighbour interactions
    
    neighbors_df_unsorted = pd.DataFrame({'neighbor_id' : user_index, 'similarity' : user_similarity_scores})
    
    
    data_user_arc = df.groupby('user_id').size().reset_index(name = 'number_articles')

    dict_users = data_user_arc.set_index('user_id')['number_articles'].to_dict()
    
    

    neighbors_df_unsorted['num_interactions'] = neighbors_df_unsorted['neighbor_id'].apply(lambda  x: dict_users[x])
    
    neighbors_df_unsorted = neighbors_df_unsorted.loc[neighbors_df_unsorted['neighbor_id'] != user_id]   
    
    
    #final neighbour sorted first on similarity and next on number of interactions
    neighbors_df = neighbors_df_unsorted.sort_values(['similarity', 'num_interactions'], ascending = False).reset_index()
    
    
    
    return neighbors_df # Return the dataframe specified in the doc_string


def user_user_recs_part2(user_id, m=10):
    '''
    INPUT:
    user_id - (int) a user id
    m - (int) the number of recommendations you want for the user
    
    OUTPUT:
    recs - (list) a list of recommendations for the user by article id
    rec_names - (list) a list of recommendations for the user by article title
    
    Description:
    Loops through the users based on closeness to the input user_id
    For each user - finds articles the user hasn't seen before and provides them as recs
    Does this until m recommendations are found
    
    Notes:
    * Choose the users that have the most total article interactions 
    before choosing those with fewer article interactions.

    * Choose articles with the articles with the most total interactions 
    before choosing those with fewer total interactions. 
   
    '''
    # Your code here
    
    recs = []
    
    rec_names = []
    
    #obtaining articles already seen by the user id to avoid recommending it
    articles_ids_seen, articles_names_seen  = get_user_articles(user_id, user_item=user_item)
    
    #obtaining the sorted neighbours to the specific user id
    neighbors_ids = get_top_sorted_users(user_id, df=df, user_item=user_item)
    
    #list of neighbours
    closest_neighbors = list(neighbors_ids['neighbor_id'])
    
    #obtaining number of interactions for articles
    article_interactions = pd.DataFrame(df.groupby(['article_id', 'title']).size().reset_index(name = 'interactions'))
        
    #sorting the articles based on number of interactions    
    article_interactions_sorted = article_interactions.sort_values('interactions', ascending = False).reset_index()
    
    
    #top sorted article ids
    ranked_articles_ids = np.array(list(article_interactions_sorted['article_id'])).ravel()
    
    #top sorted article names
    ranked_articles_names = np.array(list(article_interactions_sorted['title'])).ravel()
    
    
    
    for neighbor in closest_neighbors:
        
        #obtaining neighbour articles ids and names
        
        articles_ids_neighbor, articles_names_neighbor = get_user_articles(neighbor, user_item=user_item)
        
        #article ids not seen, neighbour has seen
        newrecs_ids_unsorted = np.setdiff1d(np.array(articles_ids_neighbor), np.array(articles_ids_seen), assume_unique=True)
        
        #article names not seen, neighbours has seen
        newrecs_names_unsorted = np.setdiff1d(np.array(articles_names_neighbor), np.array(articles_names_seen), assume_unique=True)
        
        #Recommending the article ids based on the ranked articles in the database
        newrecs_ids_sorted = list(ranked_articles_ids[np.in1d(ranked_articles_ids, newrecs_ids_unsorted)])
        
        #Recommending the article names based on the ranked articles in the database
        newrecs_names_sorted = list(ranked_articles_names[np.in1d(ranked_articles_names, newrecs_names_unsorted)])
    
        
        #concatenate recommendations
        recs = np.concatenate([newrecs_ids_sorted, recs], axis=0)
        
        rec_names = np.concatenate([newrecs_names_sorted, rec_names], axis=0)
        
        #recommedning unique articles
        _, idx =np.unique(recs, return_index = True)
        
        #making recommendations
        recs = recs[np.sort(idx)]
        
        rec_names = rec_names[np.sort(idx)]
        
        
        if len(recs) > m:
            
            break
    
    
    
    
    
    return recs[:m], rec_names[:m]
```

```python
#Creating dataframe for content and tokenizing words
#using stemming and lemmitization for text standardization

def preprocess_features(df_content_recommender):
    
    word_reco = df_content_recommender['doc_full_name']

    stemmed = []

    stemmed_docs = []

    for sentence in word_reco:
    
        tokenize = word_tokenize(sentence)
    
        for token in tokenize:
        
            stemmed.append(porter.stem(token))
        
        stemmed_docs.append(stemmed)
    
        stemmed = []
    
    return stemmed_docs
    
    
def preprocessed_dataframe (content):
    
    #stemmed_docs = preprocess_features(content)

    processed = []

    for doc in content:
        
        sentence_joined = " ".join(doc)
        
        processed.append(sentence_joined)

    word_reco_processed = pd.Series(v for v in processed)
    
    return word_reco_processed

```

```python
def create_features(content, df_content_recommender):
    
    #creating tokenized features for each article ids/name and concanetating it to the content based dataframe
    
    count_vector = CountVectorizer(stop_words = 'english', binary = 'boolean')

    word_array = count_vector.fit_transform(content)

    features = count_vector.get_feature_names()
    
    word_array = word_array.toarray()

    df_content_features = pd.DataFrame(word_array, columns = features)
    
    df_content_recommender = df_content_recommender.reset_index(inplace = False, drop =True)
    
    df_content_system = pd.concat([df_content_recommender, df_content_features], axis = 1)

    return df_content_system
    
```

```python
def content_similarity(content, article_id):
    
    #calculating content based recommendation system using dot product similarity

    user_content = np.array(content.iloc[article_id, 2:])
    
    all_articles = np.array(content.iloc[:, 2:])
    
    dotprod = np.dot(user_content, all_articles.T)
    
    dotsort = list(dotprod.argsort()[::-1])
    
    dotsort.remove(article_id)
    
    return dotsort
    
    
    

def find_similar_articles(df_content_recommender, df_content, article_id, m):
    '''
    INPUT
    movie_id - a movie_id 
    OUTPUT
    similar_movies - an array of the most similar movies by title
    
    '''
    # find the row of each movie id
    #using on our functions above to find simialr articles for content-based system
    
    stemmed_docs = preprocess_features(df_content_recommender)
    
    preprocessed = preprocessed_dataframe(stemmed_docs)
    
    content_system = create_features(preprocessed, df_content_recommender)
    
    article_idx = np.where(content_system['article_id'] == article_id)[0][0]
    
    similar_idxs = content_similarity(content_system, article_idx)
    
    df_content = df_content.reset_index(inplace = False, drop = True)
    # pull the article titles based on the indices
    similar_articles = np.array(df_content.iloc[similar_idxs, ]['doc_full_name'])
    
    return similar_articles[:m]
```
